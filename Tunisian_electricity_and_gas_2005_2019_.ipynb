```python
#i will use the strategy mode to handle missing values because  it is used  for categorical data or for numeric data that behaves categorically, i.e., representing discrete values with frequent repeats.
#and as i observed the numericals data  repeats frequently
df["tarif_type"]=df['tarif_type'].fillna(df["tarif_type"].mode()[0])
df.loc[25769,'tarif_type']
```

```python
df["counter_number"]=df['counter_number'].fillna(df["counter_number"].mode()[0])
df["counter_statue"]=df['counter_statue'].fillna(df["counter_statue"].mode()[0])
df["counter_code"]=df['counter_code'].fillna(df["counter_code"].mode()[0])
df["reading_remarque"]=df['reading_remarque'].fillna(df["reading_remarque"].mode()[0])
df["counter_coefficient"]=df['counter_coefficient'].fillna(df["counter_coefficient"].mode()[0])
df["consommation_level_1"]=df['consommation_level_1'].fillna(df["consommation_level_1"].mode()[0])
df["consommation_level_2"]=df['consommation_level_2'].fillna(df["consommation_level_2"].mode()[0])
df["consommation_level_3"]=df['consommation_level_3'].fillna(df["consommation_level_3"].mode()[0])
df["consommation_level_4"]=df['consommation_level_4'].fillna(df["consommation_level_4"].mode()[0])
df["old_index"]=df['old_index'].fillna(df["old_index"].mode()[0])
df["new_index"]=df['new_index'].fillna(df["new_index"].mode()[0])
df["months_number"]=df['months_number'].fillna(df["months_number"].mode()[0])
df["counter_type"]=df['counter_type'].fillna(df["counter_type"].mode()[0])
```
